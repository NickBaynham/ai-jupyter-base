# Example 2 - Structured Input (Multi-Part)


```python
from jupyter_base import OpenAIClient

client = OpenAIClient()
response = client.responses.create(
    model="gpt-4.1-mini",
    instructions="You are a coding tutor.",
    input=[
        {"role": "user", "content": "What is recursion?"},
        {"role": "user", "content": "Give a Python example."}
    ]
)

print(response.output_text)
```
