### Step 1: Setup

```python
import openai
from jupyter_base import OpenAIClient
```

```python
client = OpenAIClient()
```

### Zero-Shot Prompting
In zero shot we don't provide any examples.

```python
prompt = """
  What is the sentiment of this customer's feedback?

  Customer feedback: The product works as expected.
"""
```

```python
print(f"{prompt}")
response = client.responses.create(model="gpt-4.1-mini", input=prompt)
print(response.output[0].content[0].text)
```

```python
prompt_2 = """
Role: you are a sentiment feedback rater.
Task: Classify the sentiment of this customer's feedback
Format: Respond with exactly one word: positive, negative, or neutral
Constraint: Do not provide explanations or text
Example:
Feedback: This product sucks.
negative

Customer feedback: The products is satisfactory.
"""
```

```python
print(client.responses.create(model='gpt-4.1-mini', input=prompt_2).output[0].content[0].text)
```

### One-Shot Prompting
We provide exactly one example.

```python
prompt_3 = """
Role: you are a sentiment feedback rater.
Task: Classify the sentiment of this customer's feedback
Format: Respond with exactly one word: positive, negative, or neutral
Constraint: Do not provide explanations or text
Example:
Feedback: This product sucks.
negative

Customer feedback: The products is satisfactory.
"""
print(client.responses.create(model='gpt-4.1-mini', input=prompt_3).output[0].content[0].text)
```

###  Few-Shot Prompting
Usually 3 - 5 examples to maximize accuracy
We provide more than one example.

```python
prompt_4 = """
Role: you are a sentiment feedback rater.
Task: Classify the sentiment of this customer's feedback
Format: Respond with exactly one word: positive, negative, or neutral
Constraint: Do not provide explanations or text
Examples:
Feedback: This product sucks.
negative
Feedback: The product performs it main function.
neutral
Feedback: The product is outstanding.
positive

Customer feedback: {
"""
print(client.responses.create(model='gpt-4.1-mini', input=prompt_3).output[0].content[0].text)
```

```python
feedback_cases = [
    "The product works as expected.",
    "The product was not very good.",
    "The product exceeded all my expectations."
]
for case in feedback_cases:
    prompt = f"""
    Classify the sentiment of the customer's feedback.
    Example:
    Feedback: This product sucks.
    negative

    Now classify this: Feedback: {case}
    """
    print(client.responses.create(model='gpt-4.1-mini', input=prompt).output[0].content[0].text)
```
