### Using a Function to Request GPT

```python
from jupyter_base import OpenAIClient
import openai
client = OpenAIClient()
```

```python
def ask_openai(prompt, model="gpt-4.1-mini"):
    """
       Ask OpenAI a question using the global client
    """
    try:
        response = client.responses.create(
            model = model,
            input = prompt
        )
        return response.output_text.strip()
    except openai.AuthenticationError:
        return "ERROR: Invalid API Key. Check your .env file"
    except openai.RateLimitError:
        return "ERROR: Rate Limit reached. Wait and try again."
    except openai.APIConnectionError:
        return "ERROR: Network issue. Check your internet."
    except openai.BadRequestError:
        return "ERROR: Bad request. Check model name."
    except Exception as e:
        return f"API ERROR: {str(e)}"
    
```

```python
ask_openai("Who was Julius Ceasar?")
```
