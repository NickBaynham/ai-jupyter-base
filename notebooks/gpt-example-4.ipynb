### Example 4 - Passing Variables (Dynamic Prompts)

```python
from jupyter_base import OpenAIClient

client = OpenAIClient()

topic = "vector databases"

response = client.responses.create(
    model="gpt-4.1-mini",
    instructions="You are an expert in modern AI infrastructure.",
    input=f"Give a quick overview of {topic} and when to use them."
)

print(response.output_text)
```
