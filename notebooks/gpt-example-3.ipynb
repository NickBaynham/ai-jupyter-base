### Example 3 - Combining Instructions + Constraints

```python
from jupyter_base import OpenAIClient

client = OpenAIClient()
response = client.responses.create(
    model="gpt-4.1-mini",
    instructions=(
        "You are a technical writer. "
        "Respond in bullet points. "
        "Keep explanations under 100 words."
    ),
    input="What are the benefits of using Docker?"
)

print(response.output_text)
```
