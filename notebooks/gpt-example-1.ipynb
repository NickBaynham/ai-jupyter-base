### Example 1
Using input+instructions

```python
from jupyter_base import OpenAIClient

client = OpenAIClient()

# Simple single-turn reply (string in, string out)
text = client.complete_text(
    user="Summarize JupyterLab in one sentence.",
    model="gpt-4o-mini",
    system="You are a concise assistant.",
)
print(text)
```

```python
from jupyter_base.clients.openai_client import OpenAIClient

client = OpenAIClient()

response = client.responses.create(
    model="gpt-4.1-mini",
    instructions="You are a helpful assistant that explains things clearly and concisely.",
    input="Explain how neural networks work in simple terms."
)

print(response.output[0].content[0].text)
```
